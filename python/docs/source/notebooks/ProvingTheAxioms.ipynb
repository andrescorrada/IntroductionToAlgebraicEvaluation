```python
labels = ('a','b','c')
classifiers = (1,2,3)
mIdeal = ntqr.raxioms.MAxiomsIdeal(labels,classifiers,2)
Matrix([mIdeal.m_two_ideal_agreement((1,2))[label] for label in labels])
```

```python
mIdeal.all_agree_subs_dict
```

```python
cExpr = mIdeal.m_two_ideal_agreement((1,2))['a']
cExpr
```

```python
iExpr = mIdeal.m_complex[(1,2)]["axioms"]['a']
```

```python
cExpr == iExpr
```

```python
iExpr == sympy.simplify(cExpr.subs(mIdeal.all_agree_subs_dict))
```
